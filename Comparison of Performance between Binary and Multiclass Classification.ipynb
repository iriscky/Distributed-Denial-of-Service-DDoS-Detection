## Comparison of Performance between DDoS Binary and Multiclass Classification

```python

import pickle #load saved file 
import matplotlib.pyplot as plt #interpret result
import numpy as np
```

```python
# Load binary pickle file
pickle_binary = r'Binary/Dataset/Metrics/metrics_binary.pkl'
with open(pickle_binary, 'rb') as f:
    binary = pickle.load(f)

# Load the multiclass pickle file
pickle_multiclass = r'Multiclass/Dataset/Metrics/metrics_multiclass.pkl'
with open(pickle_multiclass, 'rb') as f:
    multiclass = pickle.load(f)

plt.rcParams.update({'font.size': 25})
```

### Performance in terms of accuracy

```python

# Extract accuracy scores from both files
accuracy_scores_binary = binary['accuracy']
accuracy_scores_multiclass = multiclass['accuracy']

# Prepare data for plotting
models_binary = list(accuracy_scores_binary.keys())
scores_binary = list(accuracy_scores_binary.values())

models_multiclass = list(accuracy_scores_multiclass.keys())
scores_multiclass = list(accuracy_scores_multiclass.values())

# Set the width of the bars
bar_width = 0.4

# Set the positions of the bars on the x-axis
x_pos1 = np.arange(len(models_binary))
x_pos2 = [x + bar_width for x in x_pos1]

# Combine the x positions and labels
x_combined = np.concatenate((x_pos1, x_pos2))
models_combined = models_binary + models_multiclass

# Plot the bar graph
plt.figure(figsize=(25, 12))
plt.bar(x_pos1, scores_binary, width=bar_width, align='center', alpha=1, label='Binary Classification', zorder=3)
plt.bar(x_pos2, scores_multiclass, width=bar_width, align='center', alpha=1, label='Multiclass Classification', zorder=3)

# Set the x-tick positions and labels
xtick_positions = x_pos1 + bar_width / 2
xtick_labels = models_binary

# Plot the x-ticks
plt.xticks(xtick_positions, xtick_labels, rotation=20, ha='right')

plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Comparison of Accuracy between Binary Classification and Multiclass Classification', fontsize=30)
plt.legend(loc='center right', bbox_to_anchor=(1.35,0.5))


# Add data labels for Model 1
for i, v in enumerate(scores_binary):
    plt.text(x_pos1[i], v, f"{v:.4f}", ha='center', va='bottom')

# Add data labels for Model 2
for i, v in enumerate(scores_multiclass):
    plt.text(x_pos2[i], v, f"{v:.4f}", ha='center', va='bottom')
    
plt.grid(True, zorder=0)
plt.tight_layout()
plt.show()
```
